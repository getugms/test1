## Segment the objects from a video

```python
predictor = SamGeo2(
    model_id="sam2-hiera-large",
    video=True,
)
```

```python
url = "https://github.com/opengeos/datasets/releases/download/videos/aral_sea.mp4"
```

```python
video_path = url
predictor.set_video(video_path)
```
